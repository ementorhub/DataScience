```python
def cleanSelector(obj, xpath = './text()'):
    clean_list = []
    for i in range(len(obj)):
        if obj[i].xpath(xpath).extract():
            clean_list.append(obj[i].xpath(xpath).extract()[0]) #wtheck is this :)) 
        else:
            clean_list.append(None)
    return clean_list
```
